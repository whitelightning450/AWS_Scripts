```python
import pandas as pd
import boto3
import numpy as np
import os
import warnings
from pathlib import Path
from progressbar import ProgressBar
warnings.filterwarnings("ignore")
#set path directory
cwd = os.getcwd()
os.chdir('..')
cwd = os.getcwd()
os.chdir('..')
```

```python
alldata = 'Data_1980-2020'
model = 'NWM_segments'
cols = ['Datetime', 'NWM_flow']

statepath = f"{cwd}/{alldata}/{model}"
states = os.listdir(statepath)

for state in states:
    print(state)
    filepath = f"{cwd}/{alldata}/{model}/{state}/"
    files = os.listdir(filepath)
    os.mkdir(f"{cwd}/{alldata}/NWM-AWS/{state}")
    pbar = ProgressBar()    
    for file in pbar(files):
        name = f"NWM_{file[:-14]}"
        data = pd.read_csv(f"{filepath}/{file}")
        data = data[cols]
        data.reset_index(inplace = True)
        data.rename(columns = {'index': ' '}, inplace = True)
        #csvs[name] = data
        data.to_csv(f"{cwd}/{alldata}/NWM-AWS/{state}/{name}.csv")
    
```
